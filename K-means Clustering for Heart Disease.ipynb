Problem statement

Use KMeans clustering to identify heart disease

Import libraries

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

Read files

```python
data = pd.read_csv('heart_disease.csv')
submission = pd.read_csv('sample.csv')
```

Check id against submission

```python
data['exists'] = data['id'].isin(submission['id'])
data
```

```python
data['exists'].value_counts()
```

Drop rows that are false

```python
heart_data = data[data['exists'] == True] # Set heart_data df to everything in data where the "exists" column is true
heart_data
```

Check for null values

```python
heart_data.isna().sum() # Counts how many missing entries there are in the heart_data df
```

Impute values

```python
for col in heart_data:
    if heart_data[col].dtype == 'object':
        heart_data[col].fillna('Not known', inplace=True)
    if heart_data[col].dtype == 'int64':
        heart_data[col].fillna(heart_data[col].mode()[0],inplace=True)
    if heart_data[col].dtype == 'float64':
        heart_data[col].fillna(heart_data[col].mean(),inplace=True)
        
heart_data
```

```python
heart_data.isna().sum() # NA's are now gone after data cleaning
```

Drop id and exists

```python
heart_data.drop(['id', 'exists','thal'], axis=1, inplace=True) # Dropping data that's unnecessary for analysis, also dropping "thal"
heart_data
```

```python
heart_data.info()
```

Encode

```python
for col in heart_data: # Showing the datatype of every column
    print(col, heart_data[col].dtype)
```

```python
from sklearn.preprocessing import OrdinalEncoder

# Encoding the columns in heart_data so that all categorical data is transformed into ordinal data (i.e. Male & Female to 0 & 1)

enc = OrdinalEncoder()

for col in heart_data:
    if heart_data[col].dtype == 'object':
        heart_data[col] = heart_data[col].astype(str)
        heart_data[col] = enc.fit_transform(heart_data[col].values.reshape(-1,1))
        
heart_data
```

Elbow method

```python
from sklearn.cluster import KMeans

#initialize kmeans parameters
kmeans_kwargs = {
"init": "random",
"n_init": 10,
"random_state": 42,
}

#create list to hold SSE values for each k
sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
    kmeans.fit(heart_data)
    sse.append(kmeans.inertia_)

#visualize results
plt.plot(range(1, 11), sse)
plt.xticks(range(1, 11))
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
plt.show()
```

KMeans

```python
#instantiate the k-means class, using optimal number of clusters
kmeans = KMeans(init="random", n_clusters=4, n_init=10, random_state=37) # Changed cluster amount and random state

#fit k-means algorithm to data
kmeans.fit(heart_data)

#view cluster assignments for each observation
predictions = kmeans.labels_
predictions
```

Visualise clusters

```python
from sklearn.decomposition import PCA

pca = PCA(2)

pca_data = pca.fit_transform(heart_data)

x = pca_data[:,0]
y = pca_data[:,1]

plt.scatter(x, y, c=predictions, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
print(centers.shape)
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5)
```

Prepare submission

```python
submission['cluster'] = predictions
submission.to_csv('submission.csv', index=False)
submission = pd.read_csv('submission.csv')
submission
```
